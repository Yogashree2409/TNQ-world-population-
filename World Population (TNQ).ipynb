```python
Population.rename({'Density (per km²)': 'Density', 'Area (km²)': 'Area','Growth Rate' : 'Growth_Rate', 'World Population Percentage' : 'World_Population_Percentage'}, axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Growth_Rate',y='2022 Population',data=Population)
plt.title('Population last updated in 2022 vs Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Population last updated in 2022')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.distplot(Population['Density'])
plt.title('Distribution of Population Density')
plt.xlabel('Population Density')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(Population.corr(),annot=True)
plt.title('Heatmap of Variable Correlations')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='Rank',y='Growth_Rate',data=Population)
plt.title('Rank of Population vs Growth_Rate')
plt.xlabel('Rank of Population')
plt.ylabel('Growth Rate')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='Growth_Rate',y='2022 Population',palette='deep',data=Population)
plt.title('Cluster vs Population last updated in 2021 vs Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Population last updated in 2021')
plt.show()
```
